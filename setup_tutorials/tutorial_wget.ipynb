# **How to upload files into a notebook in Colab**

In this quick tutorial we will review two use cases of '!wget' that will make using Colab much easier for your projects. In a Colab notebook you can use commandline arguments in a cell if you prepend them with a '!'. We will cover two main use cases:

1. Upload a specific file
2. Efficiently upload data

## **Use wget to import specific files**
The first use case is to upload helper/utils scripts from github. With a utils script your notbook won't be cluttered with functions that are not directly insightful for your project. There is no need to upload an entire repository, we can use it by calling '!wget' followed by the direct path to the file that we want. There is one caveat to this, if we directly copy the filepath like this:
```
!wget https://github.com/itberrios/CV_tracking/tree/main/kitti_tracker/kitti_utils.py
from kitti_utils import *
```

It will download an html version of the file and give us an error when we try to import it. Instead, navigate directly to the file and right click raw, copy the link and paste it in the cell

image.png

````python
!wget https://github.com/itberrios/CV_tracking/raw/main/kitti_tracker/kitti_utils.py
from kitti_utils import *
````

````python
# check that script was imported
!ls
````

Now we will test the import with a simple function that just adds two numbers

````python
test_func(100, 200)
````

## **Use wget to obtain data**
The next use case is to obtain data for our projects. This is particulary useful since it automates a tideous proces of manually downloading data and manualy uploading it. Even if we [mount the google drive](https://colab.research.google.com/notebooks/io.ipynb) we still need to download the data and place it in our google drive. 

This method works well for smaller dataset, I've had good experiences with datasets up to a ~3GB. If your dataset is particularly large this may work, but there could be better options to explore such as Gradient which comes with MSCOCO. To get the dataset, just rightclick the button to download and copy the link path.

image.png



We will use data from the [KITTI raw city dataset](http://www.cvlibs.net/datasets/kitti/raw_data.php). You'll notice that downloading this dataset is fairly quick.

````python
!wget https://s3.eu-central-1.amazonaws.com/avg-kitti/raw_data/2011_09_26_drive_0009/2011_09_26_drive_0009_sync.zip
````

Then we just unzip the files and place them where we want. If you click on the folder icon in the menu to left, you should notice that the .zip files was downlaoded directly to our working directory. We can unzip the file and specify the path with 'unzip', or we can unzip to the default location with either 'unzip' or 'jar xf'.

```
!unzip path/to/zipfile -d path/to/file
```
```
!jar xf path/to/zipfile
```
