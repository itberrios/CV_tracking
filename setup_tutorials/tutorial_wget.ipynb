# **How to upload files into a notebook in Colab**

In this quick tutorial we will review two use cases of '!wget' that will make using Colab much easier for your projects. In a Colab notebook you can use commandline arguments in a cell if you prepend them with a '!'

The first use case is to upload helper/utils scripts from github. With a utils script your notbook won't be cluttered with functions that are not directly insightful for your project. There is no need to upload an entire repository, we can use it by calling '!wget' followed by the direct path to the file that we want. There is one caveat to this, if we directly copy the filepath like this:
```
!wget https://github.com/itberrios/CV_tracking/tree/main/kitti_tracker/kitti_utils.py
from kitti_utils import *
```

We will donwload the file as an html and get an error when we try to import it. Instead, be sure to go to the file and right click raw, copy the link and paste it in the cell

image.png

````python
!wget https://github.com/itberrios/CV_tracking/raw/main/kitti_tracker/kitti_utils.py
from kitti_utils import *
````

The next use case is to obtain data for our projects. This is particulary useful since it automates a tideaous proces of manually downloading data and manualy uploading it. Even if we [mount the google drive](https://colab.research.google.com/notebooks/io.ipynb) we still need to download the data and place it in our google drive. 

This method works well for smaller dataset, I've had good experiances with datasets up to a ~3GB. If your dataset is particularly large this may work, but there could be better options. To get the datalink, just rightclick the button to download and copy the link path.

We will use data from the KITTI raw city dataset.

````python
!wget https://s3.eu-central-1.amazonaws.com/avg-kitti/raw_data/2011_09_28_drive_0002/2011_09_28_drive_0002_sync.zip
````

Then we just unzip the files and place them where we want. If you click on the folder icon in the menu to left, you should notice that the .zip files was downlaoded directly to our working directory. The cell below shows how we can unzip the file and specify the path, but for this example we will just unzip to the default directory

````python
#!unzip 2011_09_28_drive_0002_sync.zip -d path/to/file
````

````python
!unzip 2011_09_28_drive_0002_sync.zip 
````

This usually takes a couple of minutes total to donwload, unzip, and show up in the location, but it is much faster than manually downloading it.
